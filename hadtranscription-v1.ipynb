```python
from google.cloud import speech
import io
from pydub import AudioSegment
import os
```

```python


def transcribe_audio(audio_file):
    # Load MP3 audio file
    audio = AudioSegment.from_mp3(audio_file)

    # Convert audio to WAV format (16 kHz sample rate)
    audio = audio.set_frame_rate(16000).set_channels(1)
    wav_path = "temp_audio.wav"
    audio.export(wav_path, format="wav")

    # Read the WAV audio file
    with io.open(wav_path, "rb") as audio_file:
        content = audio_file.read()

    # Initialize Speech-to-Text client
    os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "/kaggle/input/json-key/concise-hue-414010-0ca7d551bc7f.json"  # Replace with your service account key file path
    client = speech.SpeechClient()

    # Specify audio file type and sample rate hertz
    audio = speech.RecognitionAudio(content=content)
    config = speech.RecognitionConfig(
        encoding=speech.RecognitionConfig.AudioEncoding.LINEAR16,
        sample_rate_hertz=16000,
        language_code="en-US",
    )

    # Perform the transcription
    response = client.recognize(config=config, audio=audio)

    # Output transcription
    for result in response.results:
        print("Transcript: {}".format(result.alternatives[0].transcript))

    # Clean up temporary WAV file
    os.remove(wav_path)


```

```python
# Usage
transcribe_audio("/kaggle/input/test-audio/Record (online-voice-recorder.com).mp3")
```
