```python
from transformers import pipeline
```

```python
# Load the text summarization pipeline
summarizer = pipeline("summarization")
```

```python
# Define your input text
input_text = """
    The patient is a 55-year-old male who presents with a history of chest pain and shortness of breath. He has a past medical history significant for hypertension and hyperlipidemia. On physical examination, his blood pressure is elevated at 160/90 mmHg, and he appears to be in moderate distress. Auscultation reveals diminished breath sounds in the left lower lung field. An electrocardiogram (ECG) shows ST-segment elevation in leads V2-V6. Troponin levels are elevated at 0.12 ng/mL. Based on these findings, the patient is diagnosed with an acute myocardial infarction and is scheduled for urgent cardiac catheterization.

"""
```

## Actual text    
    The patient is a 55-year-old male who presents with a history of chest pain and shortness of breath. He has a past medical history significant for hypertension and hyperlipidemia.
```python
# Generate a summary
summary = summarizer(input_text)
```

```python
# Print the summary
print(summary)
```
